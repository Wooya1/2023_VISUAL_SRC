```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import matplotlib
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic_corr=titanic.corr()
```

```python
fig, ax=plt.subplots(figsize=(7,7))
mask=np.zeros_like(titanic_corr, dtype=np.bool) # 삼각형 마스크 만들기 위해 corr 사각형  flase
mask[np.triu_indices_from(mask)]=True #위쪽 삼각형에 True
#mask=np.ones_like(corr, dtype=np.bool) # 삼각형 마스크 만들기 위해 corr 사각형  True
#mask[np.triu_indices_from(mask)]=False # 위쪽 삼각형 삼각형에 False

df_heatmap = sns.heatmap(
    titanic_corr, 
    cbar_kws={'shrink':.5}, # 컬라바 크기 절반으로 줄이기
    annot=True, 
    mask=mask, #표시하지 않는 마스크 부분 지정
    linewidth=.5, # 경계면 시선으로 구분하기
    cmap='RdYlBu_r', #red, yellow, blue 색상으로 표시
    # vmin=0.92,vmax=1
    )
```
