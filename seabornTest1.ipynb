# 1. countplot()
- 항목의 갯수를 세어줌
- 알아서 column을 구성하고 있는 value들을 구분하여 보여줌

```python
sns.countplot(x='class',data=titanic, hue='who')
plt.show()
```

```python
sns.countplot(y='class',data=titanic, hue='who')
plt.show()
```

```python
sns.countplot(x='class', data=titanic, hue='who', palette='Accent_r')
```

```python
sns.countplot(x='class', data=titanic, hue='who')
```

```python

sns.set_palette('gray')
sns.countplot(x='class', data=titanic, hue='who')
```

```python
sns.palplot(sns.color_palette("terrain_r", 7))
plt.show()
```

```python
color=['Blue','White','Red','Yellow','Green','Grey']
sns.set_palette('deep')
sns.countplot(x='class',hue='who',data=titanic, palette='color')
```

# 2. relplot
- 두 column간의 상관관계를 보여줌

```python
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='sex',size='size')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker', row='sex',col='time', data=tips)
```
