```python
from folium import plugins
import os
import numpy as np
```

```python
N=100
data=np.array(
    [np.random.uniform(low=35, high=38, size=N), 
     np.random.uniform(low=126, high=129, size=N)]
).T
data
```

```python
popups=[str(i+1) for i in range(N)]
m=folium.Map(location=[37.564444, 126.975676], zoom_start=6)
plugins.MarkerCluster(data, popups=popups).add_to(m)
m.save(os.path.join('data', 'plugins_1.html'))
m
```

```python
import pandas as pd
```

```python
cctv=pd.read_csv("data/CCTV_20190917.csv")
cctv
```

```python
popups=list(cctv.iloc[:,1])
popups
```

```python
counts=list(cctv.iloc[:,4])
counts
```

```python

lats=list(cctv.iloc[:,10])
longs=list(cctv.iloc[:,11])
print(lats)
print(longs)

lat_longs=np.array([lats,longs]).T
lat_longs
```

```python
print(len(popups), len(counts),lat_longs.shape)
```

```python
popups1, lat_longs1=[],[]
for i in range(len(counts)):
    if counts[i]>0:
        popups1.append(popups[i])
        lat_longs1.append(lat_longs[i])

print(len(popups1), len(lat_longs1))

lat_longs_arr=np.array(lat_longs1)
lat_longs_arr
```

```python
tooltip="클릭해주세요"
m=folium.Map(location=[35.1803305, 129.0516257], zoom_start=12)
plugins.MarkerCluster(lat_longs_arr, popups=popups1).add_to(m)
m.save(os.path.join('data', 'cctv_1.html'))
m
```

# 버스위치정보 지도 시각화

```python
from bs4 import BeautifulSoup as bs
import urllib.request as req
```

```python
busNum=504
key='bz6p9J0zjOTIbSrGsm6BH3B4i9IEW29rnGC6LNchn750ZVUpKW3HKr211MoqBZhOy02OyMRxsN60JqwPc0KpXQ=='
url='http://ws.bus.go.kr/api/rest/busRouteInfo/getBusRouteList?serviceKey=%s&strSrch=%s'%(key,busNum)
savename='data/businfo.xml'
req.urlretrieve(url, savename)
```

```python
xml=open(savename, 'r', encoding='utf-8').read()
soup=bs(xml,'xml')
soup
```

```python
item_list=soup.find_all('itemList')
busRoute_id=''
for item in item_list:
    busRouteId=item.find('busRouteId').text
    busRouteNm=item.find('busRouteNm').text

    if busRouteNm.strip()==str(busNum):
        print(busRouteId, busRouteNm)
        busRoute_id=busRouteId
print(busRoute_id)
```

```python
url='http://ws.bus.go.kr/api/rest/buspos/getBusPosByRtid?serviceKey=%s&busRouteId=%s'%(key,busRoute_id)
savename='data/buspos.xml'
req.urlretrieve(url, savename)
```

```python
xml=open(savename,'r',encoding='utf-8').read()
soup=bs(xml, 'xml')
busPos=[]
plainNos=[]
item_list=soup.find_all('itemList')
for item in item_list:
    plainNo=item.find('plainNo').text
    gpsX=item.find('gpsX').text
    gpsY=item.find('gpsY').text
    plainNos.append('plainNo')
    busPos.append([float(gpsY), float(gpsX)])
print(busPos)
```

```python
busPos_arr=np.array(busPos)
busPos_arr
```

```python
tooltip="클릭해주세요"
m=folium.Map(location=[37.461707, 126.874573], zoom_start=12)
plugins.MarkerCluster(busPos_arr, popups=plainNos).add_to(m)
m.save(os.path.join('data', 'cctv_1.html'))
m
```
